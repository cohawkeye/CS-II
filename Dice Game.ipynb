```python
import random

def diceRoll(dice):
    rollOne = []
    for i in range(dice):
        userRoll = random.randint(1, 6)
        cpuRoll = random.randint(1, 6)
        rollOne.append((userRoll, cpuRoll))
    return rollOne

user = input("What is your name? ")
#https://docs.python.org/3/library/string.html
print("Welcome "+ user[0].upper() + user[1:]+ "!")
cpu = "CPU"
leaderboard = {user: 0, cpu: 0}

dice = int(input("How many dice would you like to roll? "))
limit = int(input("What score would you like to go up to? "))

while leaderboard[user] < limit and leaderboard[cpu] < limit:
    rollOne = diceRoll(dice)
    for i in rollOne:
        user_value, cpu_value = i
        if user_value > cpu_value:
            print("You win this round!")
            leaderboard[user] += 1
        elif user_value < cpu_value:
            print("Computer Wins this round!")
            leaderboard[cpu] += 1
        else:
            print("It's a draw in this round!")

print("Game Over")
print(f"Final Scores - {user}: {leaderboard[user]}, {cpu}: {leaderboard[cpu]}")
```
